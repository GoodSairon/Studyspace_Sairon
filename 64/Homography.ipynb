```python
from tensorflow.keras.preprocessing.image import apply_affine_transform, load_img
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = load_img("cat.jpg", target_size=[600, 600])
plt.imshow(img)
```

```python
img = np.array(img)
```

```python
scaling_small = apply_affine_transform(
    img,
    zx=0.5,
    zy=0.5
)
plt.imshow(scaling_small)
```

```python
scaling_big = apply_affine_transform(
    img,
    zx=1.2,
    zy=1.2
)
plt.imshow(scaling_big)
```

```python
transformation = apply_affine_transform(
img,
shear=50
)
plt.imshow(transformation)
```

```python
rotation = apply_affine_transform(
img,
theta=45
)
plt.imshow(rotation)
```

```python
translation = apply_affine_transform(
    img,
    tx=30,
    ty=30
)
plt.imshow(translation)
```

```python
import cv2
```

```python
im_src = cv2.imread('chess_1.png')
```

```python
pts_src = np.array([[209, 79], [447, 71], [214, 220],[444, 231]])
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
im_dst = cv2.imread('chess_2.png')
```

```python
pts_dst = np.array([[217, 67],[471, 117],[222, 313],[382, 349]])
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, 
                               pts_dst,
                               )
```

```python
im_out = cv2.warpPerspective(im_src, h,
                             (im_dst.shape[1],
                              im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```

```python
h_1, status_1 = cv2.findHomography(pts_src, 
                               pts_dst,
                               cv2.RANSAC)
```

```python
im_out_1 = cv2.warpPerspective(im_src, h_1,
                             (im_dst.shape[1],
                              im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out_1)
plt.show()
```
