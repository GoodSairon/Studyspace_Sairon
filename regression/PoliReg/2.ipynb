```python
import pandas as pd
import numpy as np
from statistics import median
from sklearn.preprocessing import normalize, MinMaxScaler
```

```python
df = pd.read_csv('communities.csv', names=[i for i in range(len(np.array(df).transpose()))])
df
```

```python
dfn = df.drop(columns=[0, 1, 2, 3, 4])
dfn.replace(to_replace='?', value=0, inplace=True)
dfn
```

```python
norm1 = pd.DataFrame(normalize(dfn, axis=0))
print(norm1.head())
```

```python
norm2 = pd.DataFrame(MinMaxScaler(feature_range=(0, 5)).fit_transform(dfn))
print(norm2.head())
```
