```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt 
from PIL import Image, ImageEnhance
```

```python
def print_img(path=None):
    img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
print_img("cat.jpg")
```

```python
def save_img(write_path=None, path=None):
    cv2.imwrite(write_path, path)
```

```python
save_img("c.jpg", cv2.imread("cat.jpg"))
```

```python
def con_brg_img(path=None, contrast=None, brightness=None):
    im = Image.open(path)
    enhancer = ImageEnhance.Contrast(im)
    im_output = enhancer.enhance(contrast)
    enhancer = ImageEnhance.Brightness(im)
    im_output = enhancer.enhance(brightness)
    im_output.save('more-contrast-brightness-img.jpg')
```

```python
con_brg_img("cat.jpg", 1.5, 1)
```

```python
def brg_img(path=None, brightness=None):
    im = Image.open(path)
    enhancer = ImageEnhance.Brightness(im)
    im_output = enhancer.enhance(brightness)
    im_output.save('more-brightness-img.jpg')
```

```python
brg_img("cat.jpg", 2)
```

```python
def satur_img(path, saturation):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img[:, :, 1] += saturation
    cv2.imwrite("more-saturation-img.jpg", img)
```

```python
satur_img("cat.jpg", 10)
```

```python
def resize_img(path, width, height):
    img = cv2.imread(path)
    resize = (width, height)
    img = cv2.resize(img, resize)
    cv2.imwrite("resize-img.jpg", img)
```

```python
resize_img("cat.jpg", 1920, 1080)
```

```python
def cropp_img(path, width_s, width_e, height_s, height_e):
    img = cv2.imread(path)
    cropped_img = img[width_s:width_e, height_s:height_e]
    cv2.imwrite("cropped-img.jpg", cropped_img)
```

```python
cropp_img("cat.jpg", 200, 400, 100, 200)
```

```python
def rotation_img(path, angle):
    img = cv2.imread(path)
    (h, w) = img.shape[:2]
    center = (int(w / 2), int(h / 2))
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 0.6)
    img = cv2.warpAffine(img, rotation_matrix, (w, h))
    cv2.imwrite("rotated-img.jpg", img)
```

```python
rotation_img("cat.jpg", 30)
```
