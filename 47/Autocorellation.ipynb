```python
import pandas as pd
import matplotlib
from matplotlib import pyplot
```

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
series = pd.read_csv('AirPassengers.csv')
series.head()
```

```python
matplotlib.rcParams['figure.figsize'] = [10, 3]
series.plot()
pyplot.show()
```

```python
acf_value = acf(series.Passengers)
pacf_value = pacf(series.Passengers, nlags = 3)
```

```python
acf_value
```

```python
pacf_value
```

```python
matplotlib.rcParams['figure.figsize'] = [10, 3]
plot_acf(series.Passengers)
pyplot.show()
```

```python
plot_pacf(series.Passengers, lags =3)
pyplot.show()
```

```python
print(series.Passengers)
```

```python
import statsmodels.api as sa
```

```python
src_data_model = series.Passengers[:140]
model = sa.tsa.ARIMA(src_data_model).fit()
```

```python
model.summary()
```

```python
pred = model.predict(140,144, typ='levels')
```

```python
series.plot()
pred.plot(style='r--')
```

```python
comp = series.Passengers[139:144]
```

```python
import sklearn.metrics as metrics
```

```python
metrics.mean_squared_error(comp, pred)
```

```python
metrics.mean_absolute_error(comp, pred)
```

```python
model = sa.tsa.statespace.SARIMAX(src_data_model, 
                                  seasonal_order=(2,2,2,12)).fit()
```

```python
pred = model.predict(140,144, typ='levels')
```

```python
series.plot()
pred.plot(style='r--')
```

```python
metrics.mean_squared_error(comp, pred)
```

```python
metrics.mean_absolute_error(comp, pred)
```
